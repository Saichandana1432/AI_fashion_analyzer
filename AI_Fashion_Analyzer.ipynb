```python
# 1. Install libraries safely
!pip install -q transformers torch pillow numpy scikit-learn

# 2. Imports
import torch
from transformers import pipeline
from PIL import Image
import numpy as np
from sklearn.cluster import KMeans
import os

# 3. Analyzer Setup
class FashionAIAnalyzer:
    def __init__(self):
        print("Starting the Open-Access Fashion AI System...")
        self.classifier = pipeline("image-classification", model="wargoninnovation/wargon-clothing-classifier")

    def classify_clothing(self, image_path):
        image = Image.open(image_path).convert("RGB")
        predictions = self.classifier(image)
        # Process the list output safely
        if isinstance(predictions, list) and len(predictions) > 0:
            top_pred = predictions[0]
        else:
            top_pred = predictions
        return top_pred['label'], top_pred['score']

    def extract_dominant_color(self, image_path):
        image = Image.open(image_path).convert("RGB").resize((100, 100))
        img_array = np.array(image)
        pixels = img_array.reshape(-1, 3)
        kmeans = KMeans(n_clusters=1, random_state=42, n_init=10)
        kmeans.fit(pixels)
        # Ensure flat integers output array
        return [int(x) for x in kmeans.cluster_centers_[0]]

    def get_color_name(self, rgb):
        # Cleaned baseline color dictionary matrix
        fashion_colors = {
            "Red": [255, 0, 0],
            "Crimson/Burgundy": [128, 0, 32],
            "Black": [0, 0, 0],
            "White": [255, 255, 255],
            "Dark Gray": [169, 169, 169],
            "Light Gray": [211, 211, 211],
            "Navy Blue": [0, 0, 128],
            "Royal Blue": [65, 105, 225],
            "Beige/Satin Cream": [245, 245, 220],
            "Brown": [165, 42, 42],
            "Pink": [255, 192, 203],
            "Maroon": [115, 25, 25]
        }

        closest_name = "Unknown"
        min_dist = float('inf')

        for name, target_rgb in fashion_colors.items():
            dist = sum((c1 - c2) ** 2 for c1, c2 in zip(rgb, target_rgb))
            if dist < min_dist:
                min_dist = dist
                closest_name = name

        return closest_name

# 4. Smart File Scan and Execution
analyzer = FashionAIAnalyzer()

valid_extensions = ('.jpg', '.jpeg', '.png', '.webp')
uploaded_images = [f for f in os.listdir('.') if f.lower().endswith(valid_extensions)]

if uploaded_images:
    target_image = uploaded_images[0]
    print(f"Analyzing file: '{target_image}'...\n")

    label, confidence = analyzer.classify_clothing(target_image)
    rgb = analyzer.extract_dominant_color(target_image)
    color = analyzer.get_color_name(rgb)

    print("="*40)
    print(" FASHION TECH METADATA GENERATED")
    print("="*40)
    print(f"Detected Garment:  {label.upper()}")
    print(f"AI Confidence Score: {confidence*100:.2f}%")
    print(f"Extracted RGB:      {rgb}")
    print(f"Detected Color:     {color.upper()}")
    print("="*40 + "\n")
else:
    print("\n[Action Required] No images found. Please re-upload your sketch to the left file panel tab.")
```
